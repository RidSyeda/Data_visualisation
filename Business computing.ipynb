```python
import sqlite3

conn = sqlite3.connect("Moodb")
curr = conn.cursor()

create_locations_table = '''
CREATE TABLE IF NOT EXISTS locations  (
    location_id INT(3),
    street_address VARCHAR(50) UNIQUE NOT NULL,
    postal_code VARCHAR(10) NOT NULL,
    city VARCHAR(50) NOT NULL,
    state_province VARCHAR(50) NOT NULL,
    country VARCHAR(50) NOT NULL,
    CONSTRAINT l_lid_pk PRIMARY KEY (location_id)
);
'''

create_departments_table = '''
CREATE TABLE IF NOT EXISTS departments (
    department_id INT(4),
    department_name VARCHAR(20) UNIQUE NOT NULL,
    location_id INT(3),
    CONSTRAINT d_did_pk PRIMARY KEY (department_id),
    CONSTRAINT d_lid_fk FOREIGN KEY (location_id) REFERENCES locations (location_id)
);
'''

insert_locations_data = """
INSERT INTO locations (location_id, street_address, postal_code, city, state_province, country) VALUES
(100, '2 Nice Road', 'N2 7TH', 'London', 'Greater London', 'UK'),
(200, '23 Pretty Road', 'BS1 8FD', 'Bristol', 'Bristol County', 'UK'),
(300, '26 Great Street', 'BN1 4BF', 'Brighton', 'Sussex', 'UK'),
(400, '143 Lovely Road', 'CB1 2NV', 'Cambridge', 'Cambridgeshire', 'UK');
"""

insert_departments_data = """
INSERT INTO departments (department_id, department_name, location_id) VALUES
(10, 'IT', 100),
(20, 'Operations', 200),
(30, 'Sales', 300),
(40, 'Marketing', 200),
(50, 'Management', NULL);
"""

# Execute the SQL CREATE TABLE and INSERT statements
curr.execute(create_locations_table)
curr.execute(create_departments_table)
curr.execute(insert_locations_data)
curr.execute(insert_departments_data)

# Commit the changes
conn.commit()

# Query to retrieve departments and their locations
query = '''
    SELECT departments.department_id, departments.department_name, locations.street_address, 
    locations.city, locations.state_province, locations.country
    FROM departments
    INNER JOIN locations ON departments.location_id = locations.location_id
'''

# Execute the query
curr.execute(query)

# Fetch all rows
rows = curr.fetchall()

# Display the result
for row in rows:
    print(row)

# Close the connection
conn.close()
```
